## generate scalar field data then visualize

```python
import matplotlib.cm as cm 
n = 256 
x = np.linspace(-3., 3., n) 
y = np.linspace(-3., 3., n) 
X, Y = np.meshgrid(x, y) 

Z = X * np.sinc(X ** 2 + Y ** 2) 

plt.pcolormesh(X, Y, Z, cmap = cm.viridis) 
plt.show()
```

## negate scalar field then visualize

```python
plt.pcolormesh(X, Y, -Z, cmap = cm.viridis) 
plt.savefig('hom.png')
plt.show()
```

## original scalar field, inverted colormap

```python
plt.pcolormesh(X, Y, Z, cmap = cm.viridis_r)
plt.savefig('hom1.png')
plt.show()
```

## diff v(r(-d)) = -v(r(d))

```python
img = plt.imread('hom.png')
img1 = plt.imread('hom1.png')
diff = np.copy(img)
for t,v in np.ndenumerate(img):
    x,y,z = t
    v2 = img1[x][y][z]
    if v != v2:
        diff[x][y][z] = (v + v2)/2
    else:
        diff[x][y][z] = 1
        
plt.axis('off')
plt.imshow(diff)
```
