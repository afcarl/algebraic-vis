```python
import numpy as np
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
flip = lambda:'a' if np.random.randint(0,2) else 'b'
```

```python
data = [[i, 999-i, np.random.randn(), np.random.randn(), flip()] for i in range(1000)]
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.groupby(df[2]).count()
```

```python
df.columns = ['i', 'x', 'y', 'c']
```

```python
sns.pairplot(df, hue='c')
```

```python
scatter = sns.lmplot("x", "y", data=df, hue='c', fit_reg=False)
```

```python
scatter.savefig("0.png")
```
